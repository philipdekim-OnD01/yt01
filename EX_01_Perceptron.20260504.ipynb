# Perceptron 구현

## Perceptron 수식
>### y = $\sigma(\displaystyle\sum_{i=1}^{3} w_{i}x_{i} + b)$

## Load modules



```python
import numpy as np

print("NumPy Version :{}".format(np.__version__))
```

## Input, Weight, Bias



```python
# Input, Weight, Bias
X = np.array([1,2,3])
W = np.array([4,5,6])
B = 10
```

## Multiply and Summation

```python
# ================================
# 1. Element-wise multiply (원소별 곱셈)
# ================================

# W : 가중치(weight) 배열
# X : 입력(input) 데이터 배열
# 각각 같은 위치의 값끼리 곱함
# 예: [w1, w2, w3] * [x1, x2, x3] → [w1*x1, w2*x2, w3*x3]
mul = W * X

# 결과 출력
print("mul = W * X : {}".format(mul))


# ================================
# 2. Summation + Bias (합 + 편향)
# ================================

# np.sum(mul) : 곱한 결과들을 모두 더함
# 즉, w1*x1 + w2*x2 + w3*x3
# + B : bias(편향) 값 더하기
# → 이것이 퍼셉트론의 핵심 계산식
s = np.sum(mul) + B

# 결과 출력
print("s = {}".format(s))
```

## Activation Function

```python
# ================================
# 3. Activation Function (활성화 함수 - Step Function)
# ================================

# s > 10 : 조건 확인 (임계값 threshold = 10)
# 조건이 True이면 1, False이면 0
# np.float32(...) : 결과를 실수 형태로 변환 (0.0 또는 1.0)
# 즉, "결과가 10보다 크면 1, 아니면 0"
y = np.float32(s > 10)

# 결과 출력
print("y = {}".format(y))
```

입력값과 가중치를 곱하고 모두 더한 뒤 bias를 더하고, 그 결과가 특정 기준(10)을 넘으면 1, 아니면 0으로 판단하는 매우 단순한 신경망 모델입니다.
조금 더 직관적으로 보면 다음과 같은 구조입니다.
입력 X → 가중치 W 적용 → 모두 더함 → bias 추가 → 기준값 비교 → 최종 출력 y 생성
이 구조가 바로 딥러닝의 가장 기본이 되는 퍼셉트론이며, 이후 레이어를 여러 개 쌓고 활성화 함수를 바꾸면서 딥러닝으로 발전하게 됩니다.
